**Method That Predicts The winner Between Two Teams Using SVC**

```
def predict_winner(team1,team2):
  team1_nums = team1.drop(['team','year','leag','outstanding'],axis=1)
  team2_nums = team2.drop(['team','year','leag','outstanding'],axis=1)

  out_team1 = svclassifier.predict(team1_nums)
  out_team2 = svclassifier.predict(team2_nums)

  if out_team1>out_team2:
    return [1,0,0]
  elif out_team2>out_team1:
    return [0,0,1]
  elif team1['WinRate']>team2['WinRate']:
    return [1,0,0]
  elif team2['WinRate']>team1['WinRate']:
    return [0,0,1]
  else: return [0,1,0]

```
